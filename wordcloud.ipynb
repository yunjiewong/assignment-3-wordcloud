```python
from wordcloud import WordCloud,ImageColorGenerator
import imageio
import matplotlib.pyplot as plt
```

```python
text=open("output.txt",'r').read()
picture=imageio.imread("elephant.png")
wordcloud=WordCloud(background_color="white",scale=10,mask=picture).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
wordcloud.to_file("e.png")
```
